```python
# 1) 최소한의 깨끗한 row만 필터링
mask = (
    df["TITLE"].notna()
    & df["PRODUCT_TYPE_ID"].notna()
    & (df["BULLET_POINTS"].notna() | df["DESCRIPTION"].notna())
)

df_clean = df[mask]

print("원본 개수:", len(df))
print("필터 후 개수:", len(df_clean))
```

```python
df_sample = df_clean.sample(frac=0.1, random_state=42)  # 1% 샘플링
df_sample.to_csv("./dataset/sampled_products.csv", index=False)
```
